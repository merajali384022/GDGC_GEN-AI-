```python
from groq import Groq
import json
import time
from google.colab import userdata
```

```python
groq_2 = userdata.get('groq-2')
client = Groq(api_key=groq_2)
```

```python
MODEL_VERSATILE = "llama-3.3-70b-versatile"
```

```python
def generate_cricket_persona(player_role, country):
    MODEL_VERSATILE = "llama-3.3-70b-versatile"
    prompt = f"""
    you are an expert sports analyst and persona designer.

    Task: Generate three highly detailed cricket player personas for a {player_role} from {country}.

    REQUIRED FIELDS FOR EACH PERSONA:
    1. name, 2. age, 3. country, 4. occupation, 5. cricket_role, 6. batting_style, 7. bowling_style,
    8. experience_level, 9. personality_traits (list), 10. pain_points (list), 11. goals (list),
    12. motivations (list), 13. tech_savviness, 14. favorite_cricket_shot, 15. favorite_cricketer,
    16. weaknesses (list), 17. strengths (list), 18. jersey_number, 19. training_style,
    20. fitness_routine, 21. match_pressure_handling, 22. career_dream, 23. fan_following

    Format: Return ONLY valid JSON array of objects.

    Example Format:
    [
      {{
        "name": "virat kohli",
        "age": 37,
        "country": "India",
        "occupation": "Professional Athlete",
        "cricket_role": "All-rounder",
        "batting_style": "Right-handed",
        "bowling_style": "Right-arm medium",
        "experience_level": "International",
        "personality_traits": ["Focused", "Aggressive"],
        "pain_points": ["Struggles with short balls"],
        "goals": ["Win World Cup"],
        "motivations": ["Family legacy"],
        "tech_savviness": "Advanced",
        "favorite_cricket_shot": "Cover Drive",
        "favorite_cricketer": "Virat Kohli",
        "weaknesses": ["Spin on turning tracks"],
        "strengths": ["Power hitting", "Death bowling"],
        "jersey_number": 18,
        "training_style": "High Intensity",
        "fitness_routine": "Weight lifting and agility drills",
        "match_pressure_handling": "Ice cool under pressure",
        "career_dream": "Captaining the National Team",
        "fan_following": "1.1M on Instagram"
      }}
    ]
    """

    response = client.chat.completions.create(
        model=MODEL_VERSATILE,
        messages=[{"role": "user", "content": prompt}],
        temperature=1.0,

    )

    raw = response.choices[0].message.content
    data = json.loads(raw)

    if isinstance(data, dict):
        for key in data:
            if isinstance(data[key], list):
                return data[key]
    return data

# Generate and Display
personas = generate_cricket_persona('All-rounder', 'India')

print("\n✅ Parsed Personas:\n")
for i, p in enumerate(personas, 1):
    print("=" * 60)
    print(f"PERSONA {i}")
    print("=" * 60)
    print(f"Name: {p.get('name')}")
    print(f"Age: {p.get('age')}")
    print(f"Occupation: {p.get('occupation')}")
    print(f"Cricket Role: {p.get('cricket_role')}")
    print(f"\nPersonality Traits: {', '.join(p.get('personality_traits', []))}")
    print(f"Strengths: {', '.join(p.get('strengths', []))}")
    print(f"Weaknesses: {', '.join(p.get('weaknesses', []))}")
    print(f"Training Style: {p.get('training_style')}")
    print(f"Fitness Routine: {p.get('fitness_routine')}")
    print(f"Match Pressure Handling: {p.get('match_pressure_handling')}")
    print(f"Career Dream: {p.get('career_dream')}")
    print(f"Fan Following: {p.get('fan_following')}")
    print("\n")
```
